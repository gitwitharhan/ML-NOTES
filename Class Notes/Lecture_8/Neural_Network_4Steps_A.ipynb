**Step 1: - Data Definition**

```python
import numpy as np

# Define learnable training and validation data in two simple lines
xtrain = np.random.rand(5000, 32) * 10      #training features
xval = np.random.rand(1000, 32) * 10         #test features

ytrain = (xtrain[:, :5] * 2) + (np.random.randn(5000, 5) * 0.5)    #training labels
yval = (xval[:, :5] * 2) + (np.random.randn(1000, 5) * 0.5)       #test labels
```

```python
print(xtrain.shape)
print(ytrain.shape)
```

**Step 2:- Define the Neural Network Model**

```python
from keras import models, layers
```

```python
model = models.Sequential()

model.add(layers.Dense(20 , input_dim = 32))

model.add(layers.Dense(5))
```

```python
model.summary()
```

**Step 3 :- Compile Your Neural Network Model**

```python
from keras.optimizers import SGD

model.compile(optimizer=SGD(learning_rate=0.001) ,  loss = 'mse' , metrics = ['mae'])
```

**Step 4: - Train your Neural Network Model**

```python
model.fit(xtrain , ytrain , epochs = 10)
```
