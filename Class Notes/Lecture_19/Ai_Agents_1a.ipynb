```python
# Ask the Brain a hard math question

question = "What is 43568 multiplied by 89115?"
response = llm.invoke(question)
print(f"LLM's Answer: {response.content}")
```

```python
from langchain_core.tools import tool

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers together."""
    return a * b
```

```python
# Hand the calculator to the student
llm_with_tools = llm.bind_tools([multiply])
```

```python
question = "Calculate 50 times 173. Then divide that result by 5."
response = llm_with_tools.invoke(question)

# The LLM doesn't answer directly — it asks to use the tool!
print(f"Content: {response.content}")
print(f"Tool Calls: {response.tool_calls}")
```

```python
tool_calls = response.tool_calls
```

```python
tool_calls[0]
```

```python
if tool_calls:
  action = tool_calls[0]                        # Get the request
  func_name = action['name']                    # "multiply"
  args = action['args']                         # {'a': 34567, 'b': 89123}
  if func_name == "multiply":
    result = multiply.func(args['a'], args['b'])
    print(f"Calculated Answer: {result}")
else:
  print("AI answered directly:", response.content)
```
