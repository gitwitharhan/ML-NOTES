```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

from keras.models import Sequential
from keras.layers import Dense, SimpleRNN, Embedding
```

```python
train_sentences = [
    'The movie was truly amazing',      # Positive (1)
    'I absolutely loved the pizza',     # Positive (1)
    'The service was terrible',         # Negative (0)
    'I really hated the boring movie'   # Negative (0)
]
labels = np.array([1, 1, 0, 0])
```

```python
tokenizer = Tokenizer(num_words=100, oov_token="<OOV>")
tokenizer.fit_on_texts(train_sentences)
tokenizer.word_index
```

```python
sequences = tokenizer.texts_to_sequences(train_sentences)
print(sequences)
```

```python
padded_data = pad_sequences(sequences, maxlen=8, padding='post')
padded_data = np.array(padded_data)
print(padded_data)
```

```python
# 3. THE MODEL (Sequential)
model = Sequential()

# Layer 1: Embedding (The DNA Learner) - Converts numbers to meaning
model.add(Embedding(input_dim=100, output_dim=8))

# Layer 2: SimpleRNN (The Sequence Reader) - Reads step-by-step
model.add(SimpleRNN(16))

# Layer 3: Output (The Decision) - 0 to 1 Score
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# 4. TRAIN (Teach the brain)
model.fit(padded_data, labels, epochs=50)

# 5. PREDICT (Test it!)
test_text = ["movie was good"]
test_seq = tokenizer.texts_to_sequences(test_text)
test_pad = pad_sequences(test_seq, maxlen=8)

preds = model.predict(test_pad)
print("Prediction:", preds)  # Output: [[0.92]] (Positive!)
```
