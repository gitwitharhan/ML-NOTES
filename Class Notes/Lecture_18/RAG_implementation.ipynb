```python
# Load a free, pre-trained embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
# Convert sentences into vectors (embeddings)
sentences = [
 "The cat sat on the mat",
 "A kitten was sitting on the rug",     # Similar meaning!
 "The stock market crashed yesterday"   # Completely different
]
embeddings = model.encode(sentences)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
# Three sentences — two are similar, one is different
sentences = [
 "How do I return a product?", # Sentence A
 "What is the refund policy?", # Sentence B (similar to A)
 "What time does the store open?" # Sentence C (different topic)
]
```

```python
# Convert to embeddings
embeddings = model.encode(sentences)
print(embeddings)
```

```python
# Calculate how similar each pair is (1 = identical, 0 = unrelated)
similarity_matrix = cosine_similarity(embeddings)
print("Similarity between A and B:",
      round(similarity_matrix[0][1], 3))

print("Similarity between A and C:",
      round(similarity_matrix[0][2], 3))
```

```python
# Create a local vector database (no server needed!)
client = chromadb.Client()
```

```python
# Create a "collection" — like a table in SQL
collection = client.create_collection(name="company_docs_2")
```

```python
# Add documents to the collection
# ChromaDB automatically converts text to embeddings!
collection.add(
    documents=[
        "Employees get 18 days of earned leave per year.",
        "The office dress code is business casual on weekdays.",
        "Remote work is allowed on Fridays with manager approval.",
        "Health insurance covers the employee and up to 2 dependents.",
        "Annual performance reviews happen in December."
        ],
    ids=["doc1", "doc2", "doc3", "doc4", "doc5"] # Unique ID for each doc
    )
```

```python
# Now search by MEANING — not by keywords!
results = collection.query(
    query_texts=["How many vacation days do I get?"], # User's question
    n_results=2                                     # Return top 2 matches
    )

print(results)
```

**RAG PIPELINE**

```python
import chromadb
from sentence_transformers import SentenceTransformer
```

```python
# Our "company documents" (in real life, you'd load these from PDF/DOCX files)
documents = [
    "Employees are entitled to 18 days of earned leave per calendar year. "
    "Unused leave can be carried forward up to a maximum of 30 days.",
    "The company provides health insurance covering the employee, spouse, "
    "and up to two children. Dental and vision are covered under the premium plan.",
    "Remote work is permitted on Fridays with prior manager approval. "
    "Employees must be available on Slack between 10 AM and 4 PM.",
    "Performance reviews are conducted annually in December. "
    "Ratings are on a 5-point scale: Exceptional, Exceeds, Meets, Below, Unsatisfactory.",
    "The company reimburses up to Rs. 50,000 per year for professional "
    "certifications and approved training programs."
]
```

```python
# Create the embedding model and the vector database
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")
client = chromadb.Client()
collection = client.create_collection("hr_knowledge_base")
```

```python
# Convert documents to embeddings and store them
embeddings = embedding_model.encode(documents).tolist()
collection.add(
    documents=documents,
    embeddings=embeddings,
    ids=[f"doc_{i}" for i in range(len(documents))]
    )

print(f"Loaded {len(documents)} documents into the vector database!")
```

```python
import os
from openai import OpenAI
from google.colab import userdata

# Connect to OpenAI (or use Groq, Ollama, etc.)
llm_client = OpenAI(api_key=userdata.get('gpt'))
```

```python
def ask_rag(question):
  """The complete RAG pipeline in one function."""
  # STEP 1: RETRIEVE — Find relevant documents
  query_embedding = embedding_model.encode([question]).tolist()
  results = collection.query(
      query_embeddings=query_embedding,
      n_results=2 # Get top 2 most relevant chunks
      )
  context = "\n".join(results["documents"][0])

  # STEP 2: AUGMENT — Build the prompt with context
  augmented_prompt = f"""You are a helpful HR assistant. Answer the question
  using ONLY the context provided below. If the answer is not in the context,
  say "I don't have that information."
  Context: {context}
  Question: {question}
  Answer:"""

  # STEP 3: GENERATE — Send to the LLM
  response = llm_client.chat.completions.create(
  model="gpt-4o-mini",
  messages=[{"role": "user", "content": augmented_prompt}],
  temperature=0.9 # Low temperature = more factual
  )

  return response.choices[0].message.content

# Try it!
answer = ask_rag("How many leave days do I get per year?")
print(f"Answer: {answer}")
```
