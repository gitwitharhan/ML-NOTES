```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
tokens = clean_text.split(' ')
print(tokens)
```

```python
print(len(tokens))
```

```python
non_stop_words = []
for token in tokens:
  if token not in stopwords.words('english'):
    non_stop_words.append(token)

print(non_stop_words)
print(len(non_stop_words))
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

words = ["deliveries" , "waiting" , "delayed" , "studies"]

for word in words:
  print(f"{word} | {stemmer.stem(word)} | {lemmatizer.lemmatize(word)}")
```

**Bag of Wrords**

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

vectorizer = CountVectorizer()

corpus = [
    "Win free money",
    "Win huge money",
    "The deliveries were delayed"
]

sent_vector = vectorizer.fit_transform(corpus)
print(sent_vector.toarray())

print()

pd.DataFrame(sent_vector.toarray(),
             columns=vectorizer.get_feature_names_out())
```

**Padding**

```python
from tensorflow.keras.preprocessing.text import Tokenizer

sentences = [
    "I love AI",
    "The deliveries were delayed",
    "Win free money"
]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
word_index = tokenizer.word_index
print(word_index)

print()

sequences = tokenizer.texts_to_sequences(sentences)
print(sequences)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

padded_sequences = pad_sequences(sequences)
print(padded_sequences)
```
